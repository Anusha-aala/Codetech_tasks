```python
# ============================================
# TASK 4: OPTIMIZATION TECHNIQUE
# PRODUCTION OPTIMIZATION SYSTEM
# ============================================

# Step 1: Import libraries

import numpy as np
import matplotlib.pyplot as plt

from scipy.optimize import linprog

# ============================================
# OBJECTIVE FUNCTION
# ============================================

# Profit:
# Product A = ₹50
# Product B = ₹40

# Convert maximization to minimization

c = [-50, -40]

print("Objective Function:")
print(c)

# ============================================
# CONSTRAINTS
# ============================================

# 2A + B <= 100
# A + B <= 80

A = [
    [2, 1],
    [1, 1]
]

b = [100, 80]

print("\nConstraint Matrix:")
print(A)

print("\nConstraint Limits:")
print(b)

# ============================================
# VARIABLE BOUNDS
# ============================================

x_bounds = (0, None)

print("\nVariable Bounds Set!")

# ============================================
# SOLVE PROBLEM
# ============================================

result = linprog(
    c,
    A_ub=A,
    b_ub=b,
    bounds=[x_bounds, x_bounds],
    method='highs'
)

print("\nOptimization Completed!")

# ============================================
# RESULTS
# ============================================

print("\nOptimal Production:")

print("Product A =", result.x[0])

print("Product B =", result.x[1])

print("\nMaximum Profit =", -result.fun)

# ============================================
# VISUALIZATION
# ============================================

products = ['Product A', 'Product B']
values = result.x

plt.figure(figsize=(6,4))
plt.bar(products, values)

plt.title("Optimal Production Quantities")
plt.ylabel("Units Produced")

plt.savefig("optimization_result.png")
plt.show()

print("\nGraph Saved Successfully!")
```
